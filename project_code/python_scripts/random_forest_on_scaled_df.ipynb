```python
import numpy as np
from sklearn.metrics import root_mean_squared_error
from random_forest_on_drug import run_random_forest_on_drug
random_forest_model, X_train, X_test, X_val, y_train, y_test, y_val, x, y = run_random_forest_on_drug(lof_df, prism_df_scaled, merged_df, TARGET_DRUG)
random_forest_predictions_test = random_forest_model.predict(X_val)

accuracy_score_train = random_forest_model.score(X_train, y_train)
accuracy_score_val = random_forest_model.score(X_val, y_val)

random_forest_predictions_train = random_forest_model.predict(X_train)
forest_rmse_train = root_mean_squared_error(y_train, random_forest_predictions_train)

random_forest_predictions_val= random_forest_model.predict(X_val)
forest_rmse_val= root_mean_squared_error(y_val, random_forest_predictions_val)


baseline_prediction = np.full_like(y_val, y_train.mean())
baseline_rmse = root_mean_squared_error(y_val, baseline_prediction)


print(f"Random Forest score val: {accuracy_score_val}")
print(f"Random Forest score train: {accuracy_score_train}")

print(f"Random Forest RMSE val: {forest_rmse_val}")
print(f"Random Forest RMSE train: {forest_rmse_train}")
print(f"Random Forest RMSE baseline: {baseline_rmse}")
```
