```python
import pandas as pd
df = pd.read_csv('random_forest_results.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only gene columns
gene_columns = df.drop(['drug', 'rmse', 'baseline_rmse', 'pearson_correlation'], axis=1)

# Create the plot
plt.figure(figsize=(12, 6))
sns.histplot(data=gene_columns.values.flatten(), bins=100)
plt.yscale('log')
plt.xlabel('Gene importance value')
plt.ylabel('Count (log scale)')
plt.title('Distribution of Gene Importance Values')
plt.tight_layout()
plt.show()
```
