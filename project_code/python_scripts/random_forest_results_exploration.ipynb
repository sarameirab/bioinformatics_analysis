```python
# Get the row for POLYDATIN
drug_name = "MILADEMETAN (BRD:BRD-K00003406-001-01-9)"
polydatin_row_inital_results = random_forest_inital_results_sorted_top_20[random_forest_inital_results_sorted_top_20['drug'] == drug_name]


# Get all columns except the first 4
gene_columns_inital = polydatin_row_inital_results.drop(['drug', 'rmse', 'baseline_rmse', 'pearson_correlation'], axis=1)
# Convert to Series and sort by values
sorted_values_inital = gene_columns_inital.iloc[0].sort_values(ascending=False)

sorted_values_inital.head(100)

```

```python
num_zeros_inital = (sorted_values_inital == 0).sum()
num_zeros_inital
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 6))

# Create histogram with more bins and log scale on y-axis
plt.hist(sorted_values_inital[sorted_values_inital > 0.001].values, bins=100, edgecolor='black')
#plt.yscale('log')  # Set y-axis to log scale
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

plt.xlabel('Gene Importance Value')
plt.ylabel('Frequency (log scale)')
plt.title('Distribution of Gene Importance Values')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()



```
